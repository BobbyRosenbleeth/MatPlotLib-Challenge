### Analysis

- Capomulin and Ramican appeared to drive the largest drops in tumor volume among the drugs tested.
- Additional studies should compare potential benefits with the side effects of taking these drugs.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_count = combined_data.nunique()[0]
print(mice_count)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = combined_data.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)
duplicate = []
for number in np.arange(0, len(duplicates), 1):
    if duplicates[number] == True:
        duplicate.append(duplicates.index[number])

mouse_number = []
for row in duplicate:
    mouse_number.append(combined_data.iloc[row,0])

mouse_number[0]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

# Find duplicate mouse id rows and save to a list

duplicate_mouse_id = []

for row in np.arange(0, len(combined_data), 1):
    if combined_data["Mouse ID"][row] == mouse_number[0]:
        duplicate_mouse_id.append(combined_data.index[row])

# Output a dataframe with those rows

combined_data.iloc[duplicate_mouse_id]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_combined_data = combined_data.drop(
    labels = duplicate_mouse_id
)

new_combined_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
updated_mice_count = new_combined_data.nunique()[0]
print(updated_mice_count)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_numpy = new_combined_data.groupby("Drug Regimen").mean()
median_numpy = new_combined_data.groupby("Drug Regimen").median()
variance_numpy = new_combined_data.groupby("Drug Regimen").var()
sdev_numpy = new_combined_data.groupby("Drug Regimen").std()
sem_numpy = new_combined_data.groupby("Drug Regimen").sem()

summary_stats_numpy_df = pd.DataFrame({
    "Mean Tumor Volume" : mean_numpy["Tumor Volume (mm3)"],
    "Median Tumor Volume" : median_numpy["Tumor Volume (mm3)"],
    "Tumor Volume Variance" : variance_numpy["Tumor Volume (mm3)"],
    "Tumor Volume Std. Dev." : sdev_numpy["Tumor Volume (mm3)"],
    "Tumor Volume Std. Err." : sem_numpy["Tumor Volume (mm3)"]
})

summary_stats_numpy_df
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

mouse_B128_timepoint = []
mouse_B128_tumorvol = []
for i in np.arange(0, len(merged_new_combined_data), 1):
     if str(merged_new_combined_data["Mouse ID"][i]) == str("b128"):
        mouse_B128_timepoint.append(merged_new_combined_data["Timepoint"][i])
        mouse_B128_tumorvol.append(merged_new_combined_data["Tumor Volume (mm3)"][i])

mouse_B128, = plt.plot(mouse_B128_timepoint, mouse_B128_tumorvol)

plt.title("Capomulin Treatment of Mouse B128")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
weight = []
avg_tumor_vol = []
mouse_id = []

mean_weight = merged_new_combined_data.groupby("Mouse ID").mean()

for i in np.arange(1, len(merged_new_combined_data), 1):
    if (merged_new_combined_data["Drug Regimen"][i] == "Capomulin") and (merged_new_combined_data["Mouse ID"][i-1] != merged_new_combined_data["Mouse ID"][i]):
        weight.append(merged_new_combined_data["Weight (g)"][i])
        mouse_id.append(merged_new_combined_data["Mouse ID"][i])
        
for i in np.arange(0, len(mouse_id), 1):
    for g in np.arange(0, len(mean_weight), 1):
        if mean_weight.index[g] == mouse_id[i]:
            avg_tumor_vol.append(mean_weight["Tumor Volume (mm3)"][g])

x_values = weight
y_values = avg_tumor_vol
plt.scatter(x_values, y_values)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

print(f"The correlation between mouse weight and the average tumor volume is {round(st.pearsonr(weight,avg_tumor_vol)[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
predicted_y = []
for i in np.arange(0, len(x_values), 1):
    calculated_y = x_values[i] * slope + intercept
    predicted_y.append(calculated_y)
plt.scatter(x_values,y_values)
plt.plot(x_values, predicted_y, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
